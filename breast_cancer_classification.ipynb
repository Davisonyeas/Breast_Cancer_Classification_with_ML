### Encoding

#### Computers only understand Binary ( 0's & 1's ), However, the target variable needs to be converted to numeric variables (Binary, 0's & 1's)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_Y = LabelEncoder()
```

```python
data.iloc[:,0] = labelencoder_Y.fit_transform(data.iloc[:,0].values)
```

### Model

```python
sns.countplot(data["diagnosis"], label="count")
plt.title("Benign and Malignant Cancer Distribution", size=20)
```

```python
# Splitting the data
```

```python
data.columns
```

```python
# features
X = data.drop("diagnosis", axis=1) 
# target
y = data[["diagnosis"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
seed = 10
# test-data is 30% of the data set
test_data_size = 0.3 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_data_size, random_state=seed)
```

```python
# Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(X_train, y_train)
```

```python
print("Accuracy of Logistic Regression: ", lg.score(X_test, y_test))
```
